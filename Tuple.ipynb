```python
#creation of tuple
subtuple=("DS","DMS","SE","CO","DAA")
print(subtuple)
```

```python
#accessing of tuple
subtuple=("DS","DMS","SE","CO","DAA")
print(subtuple)
print(subtuple[2])
print(subtuple[-2])
print(subtuple[2:5])
print(subtuple[-4:-1])
```

```python
#change tuple value
subtuple=("DS","DMS","SE","CO","DAA")
print(subtuple)
s=list(subtuple)
s[1]="CCP"
subtuple=tuple(s)
print(subtuple)
```

```python
#use loop
subtuple=("DS","DMS","SE","CO","DAA")
for x in subtuple:
  print(x)
```

```python
#check item exists
subtuple=("DS","DMS","SE","CO","DAA")
key=input("enter a sub to search")
if key in subtuple:
  print("Yes,key is in the tuple")
else:
  print("key is not found")
```

```python
#length of tuple
subtuple=("DS","DMS","SE","CO","DAA")
print(len(subtuple))
```

```python
#adding
subtuple=("DS","DMS","SE","CO","DAA")
s=list(subtuple)
s.append("DC")
subtuple=tuple(s)
print(subtuple)
```

```python
#create with one item
subtuple=("MCES",)
print(type(subtuple))
```

```python
#remove
subtuple=("DS","DMS","SE","CO","DAA")
print(subtuple)
del subtuple
print(subtuple)
```

```python
#join two tuples
subtuple=("DS","DMS","SE","CO","DAA")
subtuple1=("MCES","CCP","DC")
sub_new=subtuple1+subtuple
print(sub_new)
```

```python
#count()
subtuple=("DS","DMS","SE","CO","DAA")
print(subtuple.count("DAA"))
```

```python
#index()
subtuple=("DS","DMS","SE","CO","DAA")
print(subtuple.index("DAA"))
```
